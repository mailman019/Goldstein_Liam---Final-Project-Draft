## Have Ocean Temperatures contributed to the severity of recent hurricane seasons?
By Liam Goldstein

The 2020 Hurricane Season was one of the most active hurricane seasons in recent times, with it being part of a recent surge in active hurricane seasons. Many might speculate that global warming and rising ocean temperatures are the primary factor in this recent trend. News sources also discuss the severity of some of the recent hurricane seasons. However, is the severity all that they make it out to be?

This data analysis will analyze the total cost of hurricane seasons and the Ocean Temperature Anamoly, or the deviation from the average ocean temperature as gathered by the National Hurricane Center in the United States, and will answer the question at hand.

According to seatemperature.org, the average ocean temperature in the Atlantic Ocean is 21 degrees Celsius.

## 2. Plot the data into a histogram

```python
sortcount = hurr_data.sort_values('Total_Cost', ascending = False)
labels = sortcount['Total_Cost'].keys()
vals = sortcount['Total_Cost']

fig, ax = plt.subplots(figsize=(10,10), dpi = 80, facecolor = 'w', edgecolor = 'k')
ax.bar(labels, vals)
ax.set_xlabel("Year (200X)")
ax.set_ylabel('Total Cost (US$)')
plt.xticks(rotation=90, fontsize='small')
plt.show()
```

```python
sortcount2 = oce_temp.sort_values('Ocean_Temp_Anamoly', ascending = False)
labels2 = ['Ocean_Temp_Anamoly']
vals2 = sortcount['Ocean_Temp_Anamoly']

fig, ax = plt.subplots(figsize=(10,10), dpi = 80, facecolor = 'w', edgecolor = 'k')
ax.bar(labels2, vals2)
ax.set_xlabel("Year (200X)")
ax.set_ylabel('Temperature Deviation')
plt.xticks(rotation=90, fontsize='small')
plt.show()
```
